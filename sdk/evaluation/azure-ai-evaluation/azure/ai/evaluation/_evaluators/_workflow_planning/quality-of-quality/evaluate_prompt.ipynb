```python
import pandas as pd
from eval_utils import append_mode_label_column

results_df_for_human_agreement = pd.read_csv('workflow_planning_results_k5.tsv', sep='\t')
results_df_for_human_agreement = append_mode_label_column(results_df_for_human_agreement, label_col='model_score')
```

```python
merged_df = results_df_for_human_agreement.merge(ground_truth_df[['trace_id', 'ground_truth']], on='trace_id', how='inner')
```

```python
# compute the precision, recall and f1 score for each model based on the model_score vs ground_truth
from sklearn.metrics import classification_report

print(f"Total tie labels excluded: {merged_df['model_score'].isna().sum()}")
for model in merged_df['model'].unique():
    print(f"Classification report for model: {model}")
    model_df = merged_df[merged_df['model'] == model]
    tie_count = model_df['model_score'].isna().sum()
    print(f"Excluded tie labels: {tie_count}")
    model_df = model_df.dropna(subset=['model_score'])
    if model_df.empty:
        print('No non-tie labels available for classification report.')
        continue
    print(classification_report(model_df['ground_truth'], model_df['model_score'].astype(int), zero_division=0))
```

```python
from eval_utils import build_disagreement_inventory, select_disagreement_cases

if 'results_df_for_human_agreement' in globals():
    results_for_gap = results_df_for_human_agreement.copy()
else:
    try:
        results_for_gap = pd.read_csv('workflow_planning_results_k5.tsv', sep='\t')
    except FileNotFoundError:
        if 'results_df' in globals() and isinstance(results_df, pd.DataFrame):
            results_for_gap = results_df.copy()
        else:
            raise

comparison_df, trace_disagreement_df = build_disagreement_inventory(results_for_gap, ground_truth_df)
top_disagreement_cases = select_disagreement_cases(trace_disagreement_df, top_n=5)

mismatch_inventory = (
    comparison_df.groupby('model', as_index=False)
    .agg(total_rows=('trace_id', 'size'), mismatches=('is_mismatch', 'sum'))
    .assign(mismatch_rate=lambda df: (df['mismatches'] / df['total_rows']).round(3))
    .sort_values(['mismatches', 'mismatch_rate'], ascending=False)
)
top5_disagreement_table = top_disagreement_cases[
    [
        'trace_id',
        'ground_truth',
        'models_evaluated',
        'models_predict_true',
        'models_predict_false',
        'mismatch_models',
        'inter_model_variance',
    ]
].copy()
top5_disagreement_table['inter_model_variance'] = top5_disagreement_table['inter_model_variance'].round(3)

display(mismatch_inventory)
display(top5_disagreement_table)
```

```python
import os
import pandas as pd
from eval_utils import load_ground_truth_data, compare_reasoning_with_and_without_dedup

BASE_DIR = os.path.dirname(os.path.abspath("__file__"))
gt = load_ground_truth_data(os.path.join(BASE_DIR, "ground_truth.csv"), BASE_DIR)
slice_df = pd.read_csv(os.path.join(BASE_DIR, "disagreement_regression_slice.tsv"), sep="\t")

# High-priority slice
case_ids = slice_df.loc[slice_df["rerun_priority"] == "high", "trace_id"].tolist()

# Skip traces that now trigger NOT_APPLICABLE gate
has_errors = gt.set_index("trace_id")["workflow_trace"].apply(lambda t: bool(t.get("errors")))
case_ids = [tid for tid in case_ids if not has_errors.get(tid, False)]

rerun_df = compare_reasoning_with_and_without_dedup(
    case_trace_ids=case_ids,
    ground_truth_df=gt,
    models_to_run=["gpt-4.1-mini", "gpt-4.1"],
    eval_key="workflow_planning",
)
rerun_df.to_csv(os.path.join(BASE_DIR, "slice_rerun_results.tsv"), sep="\t", index=False)

summary = (
    rerun_df.groupby(["formatter_mode", "model"])
    .agg(
        cases=("trace_id", "count"),
        matches=("matches_ground_truth", "sum"),
        accuracy=("matches_ground_truth", "mean"),
        avg_tokens=("workflow_planning_total_tokens", "mean"),
    )
    .reset_index()
)
display(summary)
```
